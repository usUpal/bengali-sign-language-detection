```python
import cv2
import os
import time
import uuid
```

```python
IMAGES_PATH = 'Tensorflow/workspace/images/collectedimagesFor3'
```

```python
# labels = ['zero', 'one','two','three','four','five','six','seven','eight','nine','time','north','south','counting']
# labels = ['zero','one','nothing']
labels = ['three']
number_imgs = 15
```

```python
for label in labels:
    !mkdir {'Tensorflow\workspace\images\collectedimagesFor3\\'+label}
    cap = cv2.VideoCapture(0)
    print(f'collecting images for {label}')
    time.sleep(5)
    for imgnum in range(number_imgs):
        ret, frame = cap.read()
        imgname = os.path.join(IMAGES_PATH, label, label+'.'+'{}.jpg'.format(str(uuid.uuid1())))
        cv2.imwrite(imgname, frame)
        cv2.imshow('frame',frame)
        time.sleep(2)
        
        if cv2.waitKey(1) & 0xFF == ord('q'):
            break
    cap.release()
```
